```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import t, norm
```

```python
x = np.linspace(-5, 5, 1001) 
```

```python
plt.grid()

for i in range(1,11):
    plt.plot(x,t(i).pdf(x), label='n = {}'.format(i))
    plt.legend()
plt.savefig("t_by_DOF.jpg",dpi=600)
```

```python
lst = []
for i in range(1,11):
    plt.grid()
    plt.plot(x,norm(0,1).pdf(x), label='normal')
    plt.plot(x,t(i).pdf(x), label='n = {}'.format(i))
    plt.plot(x,norm(0,1).pdf(x)-t(i).pdf(x), label='error = {:.4f}'.format(sum(((norm(0,1).pdf(x)-t(i).pdf(x)))**2)), color='red')
    plt.legend()
    lst.append(sum(((norm(0,1).pdf(x)-t(i).pdf(x)))**2))
    globals()["t_compare_normal_{}.jpg".format(i)] = plt.savefig("t_compare_normal_{}.jpg".format(i),dpi=600)
    plt.show()
```

```python
plt.grid()
plt.plot(range(1,11), lst, label="error rate", color="red")
plt.legend()
plt.savefig("error_rate.jpg",dpi=600)
plt.show()
```
